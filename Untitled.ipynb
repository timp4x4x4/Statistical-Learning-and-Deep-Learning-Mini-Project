```python
x_daan_df = x_train_df[x_train_df['鄉鎮市區'] == '大安區']
anchor = [x_daan_df['橫坐標'].mean(), x_daan_df['縱坐標'].mean()]
```

```python
def cal_distance(x, y, anchor):
    square_sum = (anchor[0] - x) ** 2 + (anchor[1] - y) ** 2
    return np.sqrt(square_sum)
```

```python
new_features = []
new_features.append('距離大安區')

x_train_df['距離大安區'] = x_train_df.apply(lambda row: cal_distance(row['橫坐標'], row['縱坐標'], anchor), axis=1)

# Check the new column
print(x_train_df[['橫坐標', '縱坐標', '距離大安區']].head())
```

```python
import numpy as np

def calculate_smoothness(y_values):
    # Calculate the differences between consecutive y-values
    differences = np.diff(y_values)
    
    # Calculate the standard deviation of the differences
    smoothness_score = np.std(differences)
    
    return smoothness_score

# Example usage with your values (for instance, the 'values' array from your chart data):
smoothness = calculate_smoothness(values)
print(f"Smoothness score: {smoothness}")
```

```python
# Recalculate '區域環境評估' based on new weights
def get_values(new_weights):
    for index, row in x_train_df.iterrows():
        region_value = sum(row[facility] * weight for facility, weight in new_weights.items())
        x_train_df.at[index, '區域環境評估'] = region_value
    # Create a dictionary to store values and calculate smoothness
    x_sum = x_train_df['區域環境評估']
    x = dict()

    for elem, value in zip(x_sum, y_train_df['單價元平方公尺']):
        if elem not in x:
            x[elem] = []
        x[elem].append(value)

    # Sort keys and calculate the median of corresponding '單價元平方公尺'
    ks = sorted(x.keys())
    values = [np.median(x[k]) for k in ks]
    
    return values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# 合併標準化的數值資料與已編碼的類別資料
x_df = pd.concat([x_train_df_encoded.reset_index(drop=True), x_train_df[non_cat_features].reset_index(drop=True)], axis=1)

# 準備目標變數 y (假設 y_train_df 是目標變數)
y = y_train_df['單價元平方公尺']
```

```python
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

def train_and_evaluation(x_df, y):
# 初始化各種模型
    x_train, x_val, y_train, y_val = train_test_split(x_df, y, test_size=0.1, random_state=42)

    print("訓練集 X 的形狀:", x_train.shape)
    print("測試集 X 的形狀:", x_val.shape)
    print("訓練集 y 的形狀:", y_train.shape)
    print("測試集 y 的形狀:", y_val.shape)
    
    model = Ridge(alpha=1.0)
    
    # 訓練模型
    model.fit(x_train, y_train)

    # 預測測試集
    y_pred = model.predict(x_val)

    # 計算 RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))

    print(f"RMSE: {rmse}")
    return model
```

```python
model = train_and_evaluation(x_df, y)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 假設 model 是已訓練的隨機森林模型
feature_importance = model.feature_importances_
sorted_idx = feature_importance.argsort()

# 顯示每個特徵的名稱和它的相對重要性
for idx in sorted_idx:
    print(f"{x_df.columns[idx]}: {feature_importance[idx]:.4f}")
```

```python
# Select important features based on feature importance threshold
threshold = 0.001  # You can adjust this value
important_features = [x_df.columns[i] for i in sorted_idx if feature_importance[i] > threshold]

# Use only the important features
x_df_important = x_df[important_features]

# Train and evaluate with only important features
model = train_and_evaluation(x_df_important, y)
```
